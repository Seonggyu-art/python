```python
import pandas as pd
import folium

# 데이터 로드
data = pd.read_csv("C:\\Users\\g_gyuuuuuu\\Desktop\\택시승차대 현황.csv", encoding='cp949')
taxi = data.loc[:, ["위치명", "위도", "경도"]]

# 숫자 변환
taxi["위도"] = pd.to_numeric(taxi["위도"], errors="coerce")
taxi["경도"] = pd.to_numeric(taxi["경도"], errors="coerce")
taxi = taxi.dropna(subset=["위도", "경도"])

# 지도 생성
mapo_center = [37.5665, 126.9012]
mapo_map = folium.Map(location=mapo_center, zoom_start=13)

# 반복문을 통해 마커 추가
for index, row in taxi.iterrows():
    lat = row["위도"]
    lon = row["경도"]
    name = row["위치명"]
    
    # 아이콘 추가 (파란색 택시 아이콘)
    icon = folium.Icon(color="blue", icon="fa-taxi", prefix="fa")
    folium.Marker([lat, lon], popup=name, icon=icon).add_to(mapo_map)

# 지도 저장
mapo_map.save("mapo_taxi_stand_with_icons.html")
```
