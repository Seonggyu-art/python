```python
# 실습. 사용자에게 검색어를 입력받아 검색된 뉴스의 제목과 링크 가져오기

# 내가 입력한 사항의 뉴스 여러개 따오기
```

```python

# 사용자로부터 검색어 입력받기
query = input("검색어를 입력하세요: ")  # 예: "강릉 날씨"
url = f"https://search.naver.com/search.naver?where=news&query={query}"

# 네이버 뉴스 페이지 요청
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# 뉴스 제목과 링크 추출
news_items = soup.select("a.news_tit")  # 네이버 뉴스 검색 결과의 제목 링크 클래스

# 뉴스 제목과 링크 출력
for idx, item in enumerate(news_items, 1):
    title = item.text  # 뉴스 제목
    link = item['href']  # 뉴스 링크
    print(f"{idx}. 제목: {title}") 
    print(f"   링크: {link}")
```

```python
# 실습1. 환율정보 크롤링하기
# 네이버 > 증권 > 시장지표 > 환전 고시 환율에 접속하여,
# 아래와 출력결과 이미지대로 크롤링 하세요.
```

```python
import requests
from bs4 import BeautifulSoup

# 네이버 금융 환율 페이지 URL
url = "https://finance.naver.com/marketindex/"

# HTTP 요청 보내기
response = requests.get(url)

# HTML 파싱
soup = BeautifulSoup(response.text, "html.parser")

# 환율 정보 추출
exchange_list = soup.select("ul.data_lst li")


# 결과 출력
for item in exchange_list[0:4]:
    currency_name = item.select_one("h3.h_lst").get_text()
    exchange_rate = item.select_one("span.value").get_text()
    print(f"{currency_name.split()[1]}: {exchange_rate}")  # 미국 USD 글자 중에서 영어만 뽑을라고 split 사용용
```

```python
# 실습2. 크롤링한 정보 엑셀에 저장하기
# 실습 1에서 구한 정보를 아래와 같이 엑셀에 저장해주세요
url = "https://finance.naver.com/marketindex/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# 필요한 데이터만 크롤링
data = []
for item in soup.select("ul.data_lst li")[:4]:  # 중국 CNY까지만 가져오기 (상위 4개)
    currency_name = item.select_one("h3.h_lst").get_text(strip=True)
    exchange_rate = item.select_one("span.value").get_text(strip=True)
    data.append([currency_name, exchange_rate])

# 엑셀 파일로 저장
pd.DataFrame(data, columns=["통화명", "환율"]).to_excel("1220 환율정보_중국까지.xlsx", index=False)
```
