```python
# 기본 그래프 그리기
font1 = {'family': 'fantasy', 'color': 'deeppink', 'weight': 'normal', 'size': 'xx-large'}

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])

# labelpad: lable padding(여백), loc: x - left/center/right, y - top/center/bottom, fontdict: 폰트
plt.xlabel('x', labelpad=30, loc='left', fontdict=font1)
plt.ylabel('y', loc='top')
plt.show()

# 레이블 패드라는 소리는 x 위에 띄어쓰기 같은거.
# 교체 가능 하단단
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('x', loc='right', fontdict=font1)
plt.ylabel('yyyy', loc='top')
plt.show()
```

```python
# 범례 표시하기
# plt.legend(loc=() ncol = 열의 개수)
```

```python
line1, = plt.plot([1, 2, 3], label='label1')
line2, = plt.plot([1, 4, 6], label='label2')
line3, = plt.plot([3, 6, 9], label='label3')
plt.legend(handles=[line1, line2])
plt.show()
```

```python
# 라벨 위치 변동, (,)는 handles를 사용했을 때 넣어야 함.
line1, = plt.plot([1, 2, 3], label='label1')
line2, = plt.plot([1, 4, 6], label='label2')
line3, = plt.plot([3, 6, 9], label='label3')
plt.legend(handles=[line1, line2], loc=(0,1))
plt.show()
```

```python
# 라벨 위치 변동 차이, best 넣어보기
line1, = plt.plot([1, 2, 3], label='label1')
line2, = plt.plot([1, 4, 6], label='label2')
line3, = plt.plot([3, 6, 9], label='label3')
plt.legend(handles=[line1, line2], loc=(1,1))
plt.show()
```

```python
import numpy as np
x = np.arange(3)
y = [100,400,900]
plt.bar(x, y, color = ['r','g','b'], width = 0.7)  #width는>
plt.plot(x,y) # 선 그래프까지 만들기
plt.xticks(x, ['2018', '2019', '2020'])
# plt.yticks(y, ['a', 'b','c'])
plt.show()
```

```python
#가로로 그래프 만들어보기, bar 가 아니라 barh 임.
# width와 height 헷갈리지 말것.
y = np.arange(3)
x = [100,400,900]
plt.barh(y, x, color = ['r','g','b'], height = 0.7)
plt.plot(x,y) # 선 그래프까지 만들기
plt.yticks(y, ['2018', '2019', '2020'])
# plt.yticks(y, ['a', 'b','c'])
plt.show() 
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = 50    # 점의 개수는 50개로 선정정
x = np.random.rand(n)
y = np.random.rand(n)

area = (30 * np.random.rand(n)) ** 2  # 0과 30 사이 난수
colors = np.random.rand(n)

# cmap: viridis(default), Spectral, plasma, inferno, magma, cividis, ...
plt.scatter(x, y, s=area, c=colors, cmap='Spectral', alpha=0.5) 
# s는 마커의 크기
# c는 마커의 색상
# cmap은 컬러 맵
# alpha는 투명도
plt.show()
```

```python
# 파이차트 그리기
```

```python
ratio = [24,28,12,36]
labels = ['A', 'B', 'C', "D"]
plt.pie(ratio, labels=labels)
plt.show()
# • labels : 레이블
# • autopct : 소수점 자리
# • explode : 부채꼴이 파이 차트의 중심에서 벗어나는 정도
# • colors : 색
# • wedgeprops : 부채꼴 스타일
```

```python

ratio = [24,28,12,36]
labels = ['A', 'B', 'C', "D"]
wedgepros = {'width' :0.7,'edgecolor' : 'k'}
plt.pie(ratio, labels=labels,wedgeprops=wedgepros)
plt.show()
```

```python
ratio = [24,28,12,36]
labels = ['A', 'B', 'C', "D"]
wedgepros = {'width' :0.7,'edgecolor' : 'k'}
plt.pie(ratio, labels=labels, wedgeprops=wedgepros, autopct='%.1f%%')
plt.show()
```

```python
ratio = [24,28,12,36]
labels = ['A', 'B', 'C', "D"]
wedgepros = {'width' :0.7,'edgecolor' : 'k'}
plt.pie(ratio, labels=labels, wedgeprops=wedgepros, autopct='%.1f%%', explode= [0,0.1,0,0.1])
plt.show()
```

```python
ratio = [24,28,12,36]
labels = ['A', 'B', 'C', "D"]
wedgepros = {'width' :0.7,'edgecolor' : 'k'}
plt.pie(ratio, labels=labels, wedgeprops=wedgepros, autopct='%.1f%%', explode= [0,0.1,0,0.1], colors=['r', 'y', 'g', 'c'])
plt.show()
```

```python
ratio = [24,28,12,36]
labels = ['A', 'B', 'C', "D"]
wedgepros = {'width' :0.7,'edgecolor' : 'k'}
plt.pie(ratio, labels=labels, wedgeprops=wedgepros, autopct='%.1f%%', explode= [0,0.1,0,0.1], colors=['r', 'y', 'g', 'c'], counterclock = False)
plt.show()
```

```python
ratio = [24,28,12,36]
labels = ['A', 'B', 'C', "D"]
wedgepros = {'width' :0.7,'edgecolor' : 'k'}
plt.pie(ratio, labels=labels, wedgeprops=wedgepros, autopct='%.1f%%', 
        explode= [0,0.1,0,0.1], colors=['r', 'y', 'g', 'c'], counterclock = False, ) 
plt.show()
```

```python
# 실습 1 아래의 자료를 가지고 그래프를 그려보기
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales_2019 = [100,120,140,110,130,150,160,170,180,200,190,210]
sales_2020 = [90,110,130,120,140,160,170,160,150,180,200,190]

plt.plot(months,sales_2019, label = '2019')
plt.plot(months,sales_2020, label = '2020')
plt.xlabel('Month')
plt.ylabel('Sales')

plt.title("Monthly Sales Comparison(2019-2020)")
plt.legend()  # 범례 

plt.show()
```

```python
# 실습 2 아래의 자료를 가지고 그래프 그려보기
categories = ['Category 1','Category 2','Category 3','Category 4','Category 5']
data = [20, 35, 15, 27, 45]


plt.bar(categories, data)
plt.xticks(rotation = 45) # x축의 각도를 맞출 수 있음
plt.title("Bar chart")

plt.xlabel("Categories")
plt.ylabel("Value")
plt.grid(True, axis = 'both')

plt.ylim(0,60) # y축의 리미트를 걸 수 있음
plt.show()
```

```python
# 실습 3. 아래의 자료를 통해 그래프 그리기
labels = ['Apples', 'Banana', 'Mango', "Blueberry"]
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)


plt.pie(sizes, labels=labels, shadow = True, 
        explode = [0, 0.1, 0, 0], autopct='%.1f%%',
           startangle = 90)
plt.show()
```

```python
# 실습4. sin & cos 함수 그리기
x = np.linspace(0, 10, 100)
# sin과 cos 함수의 값을 계산
y_sin = np.sin(x)
y_cos = np.cos(x)


import numpy as np
import matplotlib.pyplot as plt

plt.plot(x, y_sin, label = 'sin(x)', color='blue', linestyle= '-', linewidth=2)  # linewidth는 선 두께 
plt.plot(x, y_cos, label = 'cos(x)', color='red', linestyle= '-', linewidth=2)

plt.title("Sine and Cosine Fuctions")
plt.grid(True, axis = 'both')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()

```
